```python
import urllib, yfinance as yf, pandas as pd, datetime, io, numpy as np, matplotlib.pyplot as plt, re
from pathlib import Path
from email.parser import BytesParser
from email import policy

size, eps, dt0 = 10, 1.0e-10, datetime.date(2004, 1, 1)

SUFFIX = {"NASDAQ":"", "NYSE":"", "AMEX":"", "NYSEARCA":"", "TSX":".TO", "TSXV":".V", "CSE":".CN", "NEO":".NE", "ASX":".AX", "LSE":".L", "XETR":".DE", "FWB":" .F", "MIL":".MI", "BME":".MC", "SIX":".SW", "OMXSTO":".ST", "OMXHEX":".HE", "OMXCOP":".CO", "OSL":".OL", "VIE":".VI"}
EURONEXT = {"france":".PA", "netherlands":".AS", "belgium":".BR", "portugal":".LS", "ireland":".IR"}

def tickers_yahoo(filename, cur):
    msg = BytesParser(policy=policy.default).parsebytes(Path(filename).read_bytes())
    html = next((p.get_payload(decode=True).decode(errors="ignore") for p in msg.walk() if p.get_content_type() == "text/html"), "")
    res = []
    for line in re.findall(r"<tr.*?</tr>", html, re.S):
        m = re.search(r'data-rowkey="([^:]+):([^\"]+)"', line)
        curs = re.findall(r'currency[^>]*>\s*(?:<!-- -->)?\s*([A-Z]{3})\s*</span>', line)
        if m and len(curs) >= 2 and curs[1] == cur:
            exch, tk = m.groups()
            sfx = next((v for k, v in EURONEXT.items() if k in line), None) if exch == "EURONEXT" else SUFFIX.get(exch)
            if sfx is not None: res.append(tk.replace(".", "-").replace("_", "-").rstrip("-") + sfx)
    return list(dict.fromkeys(res))

def GetHistos(codes, dt):
    f = yf.download(list(codes), period="max", group_by="ticker", auto_adjust=True).xs("Close", axis=1, level=1).astype(float)
    def filter_bad_data(s):
        bad = (s.dropna() / s.dropna().shift()).abs() > 10
        return s.mask(s.index <= bad[bad].index[-1], 0) if bad.any() else s
    return f[f.index.date >= dt].apply(filter_bad_data)

def Trace(cumul, title, legend=None, log=None):
    ax = cumul.plot(figsize=(12, 8), title=title, grid=True, logy=log or False)
    if legend: ax.legend(legend)
    plt.show()

def MomentumStrategy(f, label=""):
    ls_sig = f.columns.tolist()
    ac = np.nan_to_num(f.bfill().values)
    m = f.index.to_period("M")
    dates = list(np.flatnonzero(m[:-1] != m[1:])) + [len(f.index)-1]
    ac_sig, df_sig = ac[dates,:], f.iloc[dates].index
    get_mom = lambda data, d: ((pd.DataFrame(data)-pd.DataFrame(data).rolling(d+1).mean())/(pd.DataFrame(data)+eps)).fillna(0).to_numpy()
    s1, s2 = get_mom(ac, 240)[dates,:], get_mom(ac, 90)[dates,:]
    nt, ns = s1.shape
    w = np.zeros((nt, ns))
    for it in range(1, nt):
        x = s1[it, :]
        ratio = sum(np.sign(x))/(sum(np.sign(abs(x)))+eps)
        inds = np.argsort(x)[::-1]
        for j in range(ns):
            if s2[it, j] > 0: w[it, j] = w[it-1, j]
        for idx in inds:
            if len(np.where(w[it,:])[0]) == size: break
            if s1[it, idx] > 0 and s2[it, idx] > 0 and ratio > 0.1 and ac_sig[it, idx] != 0 and w[it, idx] == 0:
                w[it, idx] = 1.0 / size
    r = pd.DataFrame(ac_sig).pct_change().add(1).fillna(1)
    eq = (pd.DataFrame(w).shift(1).fillna(0) * r).sum(axis=1).add(1 - pd.DataFrame(w).shift(1).fillna(0).sum(axis=1)).cumprod()
    print(f"\n--- {label} Last 10m ---")
    for it in range(max(0, nt-10), nt):
        nn = np.where(w[it,:] != 0)[0]
        nn_sorted = sorted(nn, key=lambda ii: s1[it, ii], reverse=True)
        l = [[ls_sig[ii] + "/" + str(round(float(ac_sig[it, ii]),2))] for ii in nn_sorted]
        print(df_sig[it].strftime("%d/%m/%Y"), "| Eq:", round(float(eq.values[it]), 2), "| Pos:", l)
    return pd.DataFrame({"equity": eq.values}).set_index(df_sig)

# https://www.quantifiedstrategies.com/rotation-strategy-in-gold-and-sp500/
def GoldStrategy(dt):
    data = yf.download(['SPY', 'GLD'], period="max", auto_adjust=True)['Close'].dropna()
    f = data[data.index.date >= dt].resample('ME').last()
    ratio = f['GLD'] / (f['SPY'] + eps)
    ma = ratio.rolling(36).mean()
    w = pd.DataFrame(0.0, index=f.index, columns=['SPY', 'GLD'])
    w['GLD'] = np.where(ratio > ma, 1.0, 0.0)
    w['SPY'] = 1.0 - w['GLD']
    r = f.pct_change().add(1).fillna(1)
    eq = (w.shift(1).fillna(0) * r).sum(axis=1).replace(0, 1).cumprod()
    print("\n--- GOLD Last 10m ---")
    for i in range(max(0, len(f)-10), len(f)):
        pos = {k: round(float(v), 2) for k, v in w.iloc[i][w.iloc[i]>0].to_dict().items()}
        print(f.index[i].strftime("%d/%m/%Y"), "| Eq:", round(float(eq.iloc[i]), 2), "| Pos:", str(pos))
    return eq.to_frame("equity")

# https://cssanalytics.wordpress.com/2025/03/20/the-growth-and-inflation-sector-timing-model/
def SectorStrategy(dt):
    ls_sig = ['SPY', 'XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV']
    f = yf.download(ls_sig, period="max", auto_adjust=True, progress=False)["Close"].ffill().loc[lambda x: x.index.date >= dt].resample("ME").last().dropna()
    inf = (0.5*f['XLE'] + 0.17*f['XLI'] + 0.17*f['XLF'] + 0.16*f['XLB']) / (0.34*f['XLP'] + 0.33*f['XLU'] + 0.33*f['XLV'] + 1e-12)
    weights = pd.DataFrame(0.0, index=f.index, columns=f.columns)
    for l in [12, 24, 36, 48]:
        hi, bull = inf > inf.rolling(l).median(), f['SPY'] > f['SPY'].rolling(l).median()
        weights['XLE'] += np.where(hi & bull, 0.25, 0);  weights['XLV'] += np.where(hi & ~bull, 0.25, 0)
        weights['XLK'] += np.where(~hi & bull, 0.25, 0); weights['XLP'] += np.where(~hi & ~bull, 0.25, 0)
    r = f.pct_change().add(1).fillna(1)
    w_lag = weights.shift(1).fillna(0)
    eq = ((w_lag * r).sum(axis=1) + (1 - w_lag.sum(axis=1))).cumprod()
    print("\n--- SECTOR Last 10m ---")
    for i in range(max(0, len(f)-10), len(f)):
        if weights.iloc[i].sum() > 0:
            pos = {k: round(float(v/weights.iloc[i].sum()), 2) for k, v in weights.iloc[i][weights.iloc[i]>0].to_dict().items()}
        else: pos = "Cash"
        print(f.index[i].strftime("%d/%m/%Y"), "| Eq:", round(float(eq.iloc[i]), 2), "| Pos:", str(pos))
    return eq.to_frame("equity")

def BondStrategy(dt):
    ls_sig = ['SHY', 'IEF', 'LQD', 'HYG', 'TIP', 'CWB']
    df = yf.download(ls_sig, period="max", auto_adjust=True, progress=False)["Close"].ffill()
    f = df[df.index.date >= dt].resample("ME").last()
    r = f.pct_change()
    mom = (r + r.rolling(3).sum() + r.rolling(6).sum() + r.rolling(12).sum()) / 4.0
    rank = mom.rank(axis=1, ascending=False)
    weights = ((rank <= 3) & (mom > 0)).astype(float) / 3.0
    rets = f.pct_change().add(1).fillna(1)
    w_lag = weights.shift(1).fillna(0)
    eq = ((w_lag * rets).sum(axis=1) + (1 - w_lag.sum(axis=1))).cumprod()
    print("\n--- BOND Last 10m ---")
    for i in range(max(0, len(f)-10), len(f)):
        pos = {k: round(float(v), 2) for k, v in weights.iloc[i][weights.iloc[i]>0].to_dict().items()}
        print(f.index[i].strftime("%d/%m/%Y"), "| Eq:", round(float(eq.iloc[i]), 2), "| Pos:", str(pos))
    return eq.to_frame("equity")

# --- Run ---
weights_dict = {"EUR": 0.1, "USD": 0.1, "AUD": 0.1, "GBX": 0.1, "CAD": 0.1, "GOLD": 0.15, "SECT": 0.15, "BOND": 0.2}
files = {"EUR":"STOXX 600 Components – TVC_SXXP Stocks — TradingView.mhtml", "USD":"S&P 500 Index Components – SP_SPX Stocks — TradingView.mhtml", "AUD":"S&P_ASX 200 Index Components – ASX_XJO Stocks — TradingView.mhtml", "GBX":"FTSE 350 Index Components – FTSE_NMX Stocks — TradingView.mhtml", "CAD":"S&P_TSX Composite Index Components – TSX_TSX Stocks — TradingView.mhtml"}
sim_momentum = pd.concat({k: MomentumStrategy(GetHistos(tickers_yahoo(v, k), dt0), k)['equity'] for k, v in files.items()}, axis=1).ffill()
gold_eq = GoldStrategy(dt0).reindex(sim_momentum.index, method='ffill').fillna(1.0)
sect_eq = SectorStrategy(dt0).reindex(sim_momentum.index, method='ffill').fillna(1.0)
bond_eq = BondStrategy(dt0).reindex(sim_momentum.index, method='ffill').fillna(1.0)
sim = pd.concat([sim_momentum, gold_eq.rename(columns={"equity":"GOLD"}), sect_eq.rename(columns={"equity":"SECT"}), bond_eq.rename(columns={"equity":"BOND"})], axis=1).ffill().fillna(1.0)
Trace(sim, "Equities", sim.columns.tolist(), log=True)
weights_series = pd.Series(weights_dict)
weights_series = weights_series / weights_series.sum()
combo = (1 + sim.pct_change(fill_method=None).fillna(0).dot(weights_series[sim.columns])).cumprod().rename("Combo")
Trace(combo.to_frame(), "Combo", ["Combo"], log=True)
r = combo.pct_change(fill_method=None).dropna()
mdd = ((combo.cummax() - combo) / combo.cummax()).max()
print(f"\nMean: {round(100*12*float(r.mean()),2)}% | Stdev: {round(100*12**0.5*float(r.std()),2)}% | Sharpe: {round((12*float(r.mean()))/(12**0.5*float(r.std())),2)} | MaxDD: {round(100*float(mdd),2)}%")
print("\n--- Strategy Coefficients ---")
print(weights_series.to_string())
```
